```python
import pandas as pd
import plotly.express as px

df=pd.read_csv('RA2019_A2.csv')
```

```python
df.columns
```

```python
df['States/UTs'].fillna(0, inplace=True)
df['State/UT-Wise Total Number of Road Accidents during 2016'].fillna(0, inplace=True)
df['State/UT-Wise Total Number of Road Accidents during 2017'].fillna(0, inplace=True)
df['State/UT-Wise Total Number of Road Accidents during 2018'].fillna(0, inplace=True)
df['State/UT-Wise Total Number of Road Accidents during 2019 - Numbers'].fillna(0, inplace=True)
df['State/UT-Wise Total Number of Road Accidents during 2019 - Rank'].fillna(0, inplace=True)
df['Share of States/UTs in Total Number of Road Accidents - 2016'].fillna(0, inplace=True)
df['Share of States/UTs in Total Number of Road Accidents - 2017'].fillna(0, inplace=True)
df['Share of States/UTs in Total Number of Road Accidents - 2018'].fillna(0, inplace=True)
df['Share of States/UTs in Total Number of Road Accidents - 2019'].fillna(0, inplace=True)
df['Total Number of Accidents Per Lakh Population - 2016'].fillna(0, inplace=True)
df['Total Number of Accidents Per Lakh Population - 2017'].fillna(0, inplace=True)
df['Total Number of Accidents Per Lakh Population - 2018'].fillna(0, inplace=True)
df['Total Number of Accidents Per Lakh Population - 2019'].fillna(0, inplace=True)
df['Total Number of  Road Accidents per 10,000 Vehicles - 2016'].fillna(0, inplace=True)
df['Total Number of  Road Accidents per 10,000 Vehicles - 2017'].fillna(0, inplace=True)
df['Total Number of  Road Accidents per 10,000 Vehicles - 2018'].fillna(0, inplace=True)
df['Total Number of  Road Accidents per 10,000 Vehicles - 2019'].fillna(0, inplace=True)
df['Total Number of  Road Accidents per 10,000 Km of Roads - 2016'].fillna(0, inplace=True)
df['Total Number of  Road Accidents per 10,000 Km of Roads - 2017'].fillna(0, inplace=True)
df['Total Number of  Road Accidents per 10,000 Km of Roads - 2018(P)'].fillna(0, inplace=True)
```

```python
df
```

```python
df['Total_Accidents'] = df[['State/UT-Wise Total Number of Road Accidents during 2016', 'State/UT-Wise Total Number of Road Accidents during 2017', 'State/UT-Wise Total Number of Road Accidents during 2018', 'State/UT-Wise Total Number of Road Accidents during 2017', 'State/UT-Wise Total Number of Road Accidents during 2018','State/UT-Wise Total Number of Road Accidents during 2019 - Numbers']].sum(axis=1)

top_cities = df.nlargest(37, 'Total_Accidents')

for index, row in top_cities.iterrows():
    city_name = row['States/UTs']
    years = ['2016', '2017', '2018', '2019']
    accidents = row[['State/UT-Wise Total Number of Road Accidents during 2016', 'State/UT-Wise Total Number of Road Accidents during 2017', 'State/UT-Wise Total Number of Road Accidents during 2018', 'State/UT-Wise Total Number of Road Accidents during 2019 - Numbers']]

    f=px.bar(x=years, y=accidents, title=f"Accidents in {city_name}",labels={'x': 'Year', 'y': 'Number of Accidents'})
    f.show()
```

```python
f1=px.bar(df, x='States/UTs', y='State/UT-Wise Total Number of Road Accidents during 2019 - Numbers', title='Total Accidents State/UT during 2019')
f1.update_layout(yaxis_title='Number of Accidents', height=1000)
f1.show()

top=df.nlargest(37, 'State/UT-Wise Total Number of Road Accidents during 2019 - Numbers')
f2=px.bar(top, x='States/UTs', y=['State/UT-Wise Total Number of Road Accidents during 2016', 'State/UT-Wise Total Number of Road Accidents during 2017', 'State/UT-Wise Total Number of Road Accidents during 2018', 'State/UT-Wise Total Number of Road Accidents during 2019 - Numbers'], title='Comparison of Accidents in 2016 and 2019')
f2.update_layout(yaxis_title='Number of Accidents', height=1000)
f2.show()

f3=px.bar(df, x='States/UTs', y='State/UT-Wise Total Number of Road Accidents during 2019 - Rank', title='Ranking of Accidents by State in 2019')
f3.update_layout(yaxis_title='Ranking', height=1000)
f3.show()

f4=px.pie(df, names='States/UTs', values='Share of States/UTs in Total Number of Road Accidents - 2019', title='Accidents in 2019 by State')
f4.update_layout(width=800, height=600)
f4.show()
```
